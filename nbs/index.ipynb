<h3 class="card-title">Getting Over ANOVA</h3>
<p class="card-text">

<h3 class="card-title">What Is a Hungry Fly?</h3>
<p class="card-text">
Tracking thousands of <em>Drosophila</em> across millions of frames to understand how
serotonin implements hunger states. Custom hardware, computer vision, 17 behavioural
metrics, high-dimensional statistics. The hard question: how do you know if a
manipulation actually changed the state, or just one metric?
</p>
